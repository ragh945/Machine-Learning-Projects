# Text Preprocessing

```python
#User defined function for text-preprocessing
def preprocess_text(text):
    """
    Preprocess the input text by performing the following steps:
    1. Convert to lowercase
    2. Remove punctuation
    3. Remove stopwords
    4. Tokenize the text

    Args:
        text (str): The text to preprocess.

    Returns:
        str: The preprocessed text.
    """
    # Convert to lowercase
    text = text.lower()
    
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    # Tokenize
    tokens = word_tokenize(text)
    
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    
    # Join tokens back to a string
    processed_text = ' '.join(tokens)
    
    return processed_text
```

```python
#Applying the Textpreprocessing to the feature "Comments"
df1["Comments"]=df1["Comments"].apply(preprocess_text)
df1
```

# Text Vectorization

```python
#Applying the feature extraction by tfidf vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(stop_words="english")
```

```python
tfidf.fit_transform(x["Comments"]).toarray()
```

```python
X=pd.DataFrame(tfidf.fit_transform(x["Comments"]).toarray(),columns=tfidf.get_feature_names_out())
X
```

```python
#Splitting thr data into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=23)
```

# Model Building and Evaluation

# Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
NB=MultinomialNB()
NB.fit(X_train,y_train)
y_pred=NB.predict(X_test)
print(classification_report(y_pred,y_test))
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
lin=LogisticRegression()
lin.fit(X_train,y_train)
y_pred=lin.predict(X_test)
print(classification_report(y_pred,y_test))
```

# RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
Rf=RandomForestClassifier()
Rf.fit(X_train,y_train)
y_pred=Rf.predict(X_test)
print(classification_report(y_pred,y_test))
```
